```python
# PREAMBLE
obs = 'OPERA_pcp'
case = 'VALENCIA'
exp = 'VAL500m_46h1_de2'
init_time = '2022050200'
case_ini = '2022050317'
case_end = '2022050323'
case_bounds = [-5., 6.5, 35., 45.]
verif_domain = [-1.6, 0.9, 38.65, 40.65]
```

```python
import sys
sys.path.append('../libs/')
from LoadWriteData import LoadConfigFileFromYaml

obs_db, var_verif = obs.split('_')

# Load obs & exp config files
config_obs_db = LoadConfigFileFromYaml(f'../../config/config_{obs_db}.yaml')
config_exp = LoadConfigFileFromYaml(f'../../config/config_{exp}.yaml')
```

```python
# check the params
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
import cartopy.crs as ccrs
from plots import PlotMapInAxis, PlotBoundsInAxis, PlotContourDomainInAxis
from dicts import get_data_function, get_grid_function, colormaps

first_date = True
#date = datetime.strptime(case_ini, '%Y%m%d%H')
dates = pd.date_range(datetime.strptime(case_ini, '%Y%m%d%H'), datetime.strptime(case_end, '%Y%m%d%H'), freq = '1H').to_pydatetime()
for date in dates:
    if config_exp['vars'][var_verif]['accum'] == True:
        date = date + timedelta(hours = 1)
    lead_time = int((date - datetime.strptime(init_time, '%Y%m%d%H')).total_seconds()/3600.0)

    obs_file = datetime.strftime(date, f"../../OBSERVATIONS/data_{obs}/{case}/{config_obs_db['format']['filename']}")

    if first_date == True:
        obs_lat, obs_lon = get_grid_function[config_obs_db['format']['extension']](obs_file)
        exp_lats_lon_init = {}
        exp_file = f"../../SIMULATIONS/{exp}/data_orig/{init_time}/{config_exp['format']['filename'].replace('%Y%m%d%H', init_time).replace('%LLL', str(lead_time).zfill(3))}" # maybe I can not use data_orig folder... think about it
        exp_lat, exp_lon = get_grid_function[config_exp['format']['extension']](exp_file)
        first_date = False

    obs_values = get_data_function[config_obs_db['format']['extension']](obs_file, [var_verif,])

    fig = plt.figure(0, figsize = (9.0 / 2.54, 9.0 / 2.54), clear = True)
    ax = fig.add_subplot(1, 1, 1, projection = ccrs.PlateCarree())
    PlotMapInAxis(ax, obs_values, obs_lat, obs_lon, extent = case_bounds, titleLeft = obs_db, titleRight = datetime.strftime(date, '%Y%m%d%H'), cbLabel = f"{config_obs_db['vars'][var_verif]['description']} ({config_obs_db['vars'][var_verif]['units']})", cmap = colormaps[var_verif]['map'], norm = colormaps[var_verif]['norm'])
    PlotContourDomainInAxis(ax, exp_lat, exp_lon, text = init_time)
    PlotBoundsInAxis(ax, verif_domain, text = 'verif', color = 'black')
    plt.show()
```
